```python
from textgenrnn import textgenrnn
import pickle
```

```python
with open('data/cocktails_for_nn.pickle', 'rb') as fp:
    recipes = pickle.load(fp)
```

```python
cocktail_generator = textgenrnn()
```

```python
single_ingredient_inputs = []
all_ingredient_inputs = []
for ii in range(0, len(recipes)):
    for ingredient in recipes[ii][1]:
        single_ingredient_inputs.append(ingredient)
    all_ingredient_inputs.append(', '.join(recipes[ii][1]))
```

```python
len(all_ingredient_inputs)
```

```python
full_cocktail_generator = textgenrnn()
```

```python
full_cocktail_generator.train_on_texts(all_ingredient_inputs, num_epochs=35, dropout=0.10, word_level=True, 
                                  max_length=20, line_deliminted = True, rnn_bidirectional=True, rnn_size = 3, 
                                  dim_embeddings=300, batch_size=128)
```

```python
names = []
for ii in range(0, len(recipes)):
    names.append(recipes[ii][0])
names[:5]
```

```python
cocktail_name_generator = textgenrnn()
```

```python
cocktail_name_generator.train_on_texts(names, num_epochs=35, dropout=0.10, word_level=True, 
                                  max_length=6, line_deliminted = True, rnn_bidirectional=True, rnn_size = 3, 
                                  dim_embeddings=300, batch_size=128)
```

```python
cocktail_name_generator.generate()
```

```python
full_cocktail_generator.generate()
```

```python
cocktail_name_generator.save('name_weights.hdf5')
```

```python
full_cocktail_generator.save('cocktail_weights.hdf5')
```
